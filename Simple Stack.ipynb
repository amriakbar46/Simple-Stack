```python
import os
```

```python
class Stack:
    def __init__(self):
        self.tas = []
        
    #cek tas kosong
    def isEmpty(self):
        return self.tas == []
    
    #tambah barang ke tas
    def push(self, barang):
        self.tas.append(barang)
        
    #keluarin barang dari tas
    def pop(self):
        if self.isEmpty():
            print("Tas Kosong Bos,Nga ada yang mau dikeluarin")
        else :
            return self.tas.pop()
        
    #lihat isi terakhir dari tas
    def peek(self):
        if self.isEmpty():
            print("Tas kosong bos")
        else :
            return self.tas[-1]
    
    #cek ada berapa barang dalam tas
    def size(self):
        return len(self.tas)
    
    #menu
    def menu(self):
        pilihan = "y"
        while (pilihan == "y"):
            os.system("clear")
            print("++++++++++++++++++++++++++")
            print("+        Menu Tas        +")
            print("++++++++++++++++++++++++++")
            print("+ 1. Tambah barang       +")
            print("+ 2. Keluarin barang     +")
            print("+ 3. Cek tas kosong      +")
            print("+ 4. Cek barang terakhir +")
            print("+ 5. Cek jumlah barang   +")
            print("++++++++++++++++++++++++++")
            pilih = str(input(("Pilih menu nomor :")))
            if(pilih=="1"):
                os.system("clear")
                barang = str(input("Masukan barang yang ingin anda tambahkan: "))
                self.push(barang)
                print("Barang "+barang+" telah ditambahkan (tekan enter)")
                x = input("")
            elif(pilih=="2"):
                os.system("clear")
                print("Barang "+self.pop()+" dikeluarkan (tekan enter)")
                x = input("")
            elif(pilih=="3"):
                os.system("clear")
                print(self.isEmpty())
                x = input("")
            elif(pilih=="4"):
                os.system("clear")
                print("Barang terakhir: "+self.peek())
                x = input("")
            elif(pilih=="5"):
                os.system("clear")
                print("Jumlah barang di tas adalah: "+str(self.size())+"(tekan enter)")
                x = input("")
            else:
                pilih="y"  
        
```
